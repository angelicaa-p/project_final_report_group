```r
set.seed(160)

players_split <- initial_split(players_data, prop = 0.75, strata = played_hours)
players_train <- training(players_split)
players_test <- testing(players_split)

folds <- vfold_cv(players_train, v = 5, strata = played_hours)

ks <- tibble(neighbors = seq(from = 1, to = 100, by = 5))

players_gender_recipe <- recipe(played_hours ~ gender, data = players_train) |>
step_novel(all_predictors()) |>
step_dummy(all_nominal_predictors()) |>
step_zv(all_predictors()) |>          
step_scale(all_predictors()) |>
step_center(all_predictors()) 

players_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
set_engine("kknn") |>
set_mode("regression")

players_gender_results <- workflow() |>
add_recipe(players_gender_recipe) |>
add_model(players_spec) |>
tune_grid(resamples = folds, grid = ks) |>
collect_metrics() 

best_gender_k <- players_gender_results |> filter(.metric == "rmse") |> arrange(mean) |> slice(1) |> select(neighbors) |> pull()

best_gender_k

players_best_gender_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = best_gender_k) |>
set_engine("kknn") |>
set_mode("regression")

players_best_gender_workflow <- workflow() |>
add_recipe(players_gender_recipe) |>
add_model(players_best_gender_spec) |>
fit(players_train) 

players_test_gender_predictions <- predict(players_best_gender_workflow, players_test) |> bind_cols(players_test) |>
metrics(truth = played_hours, estimate = .pred) |>
filter(.metric == "rmse")
players_test_gender_predictions
```
