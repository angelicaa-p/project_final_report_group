Using the general summary to further explore individual variables:

```r
# Experience variable summary statistics
experience_categories <- players_data |> # summarizes frequency counts of each category
                            group_by(experience) |>
                            summarize(count = n()) |>
                             mutate(percentage = round((count / sum(count)) * 100, 2)) # expresses the categorial count as a percentage to the total count
experience_categories
```
